```python
from natasha import (
    Segmenter,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    
    Doc
)
segmenter = Segmenter()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)

text = 'Бледные просветы — посреди ветвей \n Хищные газеты — будничный трофей \n Раненой собакой бредит командир \n И это всё обычно называется, сынок - РЕАНИМАЦИЯ \n Толстые лошадки — сытые поля \n Сочные початки — солнечное вымя \n Ждёт за поворотом белый хоровод \n И это залихватcкое незримое кино- \n РЕАНИМАЦИЯ \n А вот отдельному солдату перестало умирать \n Ведь у него лишь только слово, только слово — но какое! - \n Стал он жрать/бухать/блевать — себя на части разбирать- \n Череп крепко разворочен, и мозги текут сквозь стены- \n Стены, стены, двери, окна — другие были рядом \n Торопливо постигали положение вещей \n Созерцая похождения Текущего Повсюду, \n По привычке продолжая ритуально повторять: \n РЕАНИМАЦИЯ. \n В раскалённом парке — летняя гроза \n Поздние подарки, наши голоса \n Где-то краем уха — духовой оркестр \n И это всё обычно называется, сынок- \n РЕАНИМАЦИЯ'
doc = Doc(text)

doc.segment(segmenter)
doc.tag_morph(morph_tagger)
doc.parse_syntax(syntax_parser)

sent = doc.sents[0]
sent.morph.print()
sent.syntax.print()
```

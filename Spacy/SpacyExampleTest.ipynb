```python
import spacy
from spacy.lang.ru import Russian
from spacy import displacy
nlp = Russian()
doc = nlp("Не вынесла душа поэта позора мелочных обид, восстал он против мнений света один, как прежде... и убит!")
print("is_word:    ", [token.is_alpha for token in doc])
print("is_punct:    ", [token.is_punct for token in doc])
print("is_num:    ", [token.like_num for token in doc])
```
